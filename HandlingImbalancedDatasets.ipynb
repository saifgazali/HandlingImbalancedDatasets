```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
ann = Sequential()

ann.add(Dense(units=128,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dropout(rate=0.3))
ann.add(Dense(units=64,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dropout(rate=0.3))
ann.add(Dense(units=32,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dropout(rate=0.3))
ann.add(Dense(units=1,activation = 'sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,Y_train,batch_size=32,epochs=100)
```

```python
ann.evaluate(X_test,Y_test)
```

```python
y_pred = ann.predict(X_test) > 0.5
```

```python
from sklearn.metrics import classification_report,confusion_matrix
cm = confusion_matrix(Y_test,y_pred)
print(cm)
```

```python
print(classification_report(Y_test,y_pred))
```

```python
# f1-score is 0 for the customers who exited so basically we are not predicting properly for any customer who left .
# our model is as good as saying whatever customer comes he is going to stay.
# hence accuracy is misleading even though it is as high as 80%
```

### Improving f1 score / Handling imbalance datasets

### 1) Undersampling

```python
df.head()
```

```python
class_0_counts, class_1_counts = df.Exited.value_counts()
```

```python
(class_0_counts, class_1_counts)
```

```python
df_class_0 = df[df['Exited'] == 0]
df_class_1 = df[df['Exited'] == 1]
```

```python
df_class_0.shape
```

```python
df_class_1.shape
```

```python
## Undersampling
df_class_0.sample(2) # Randomly selects 2 sample 
```

```python
df_class_0_under = df_class_0.sample(class_1_counts)
df_test_under = pd.concat((df_class_0_under,df_class_1),axis=0)

print(df_test_under.Exited.value_counts())
```

```python
X = df_test_under.drop(['RowNumber','Exited','CustomerId','Surname'],axis=1)
Y = df_test_under['Exited']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y)
```

```python
Y_train.value_counts() # almost equal number of 0 and 1 due to stratify set to y
```

```python
ann = Sequential()

ann.add(Dense(units=128,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dropout(rate=0.3))
ann.add(Dense(units=64,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dropout(rate=0.3))
ann.add(Dense(units=32,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dropout(rate=0.3))
ann.add(Dense(units=1,activation = 'sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,Y_train,batch_size=32,epochs=100)
```

```python
y_pred = ann.predict(X_test) > 0.9
```

```python
from sklearn.metrics import classification_report,confusion_matrix
cm = confusion_matrix(Y_test,y_pred)
print(cm)
```
